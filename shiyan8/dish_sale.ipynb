```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("dish_sale.xls")
```

```python
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.figure(figsize=(8, 4))
plt.plot(data[""], data["A"], color="green", label="A", marker="o")
plt.plot(data[""], data["B"], color="red", label="B", marker="s")
plt.plot(data[""], data["C"], color="skyblue", label="C", marker="x")
plt.legend()  # 
plt.ylabel("")
plt.show()
```
