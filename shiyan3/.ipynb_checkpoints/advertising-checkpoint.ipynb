```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("Advertising.csv", header = 0)
data.head()    #数据读入
```

```python
sns.pairplot(data
            , x_vars = ['TV', 'radio', 'newspaper']
            , y_vars = 'sales'
            , height = 7
            , aspect = 0.8
            , kind = 'reg')
plt.show()    #对数据框进行探索性分析，调用Seaborn包中的pairplot()方法，绘制TV、radio、newspaper这3个变量与sales变量之间的关系图
```

```python
#采用drop()方法删除数据框data中的Number以及sales两列返回另一个DataFrame对象Data，并显示Data数据集
```

```python
##第一步：构建特殊矩阵和目标数组
Data = data.drop(['Number', 'sales'], axis = 1)
Data.head()
```

```python
sales = data['sales']
type(sales)    #显示sales的数据类型
```

```python
sales = np.ravel(sales)    #将目标向量sales的数据转换为NumPy中的ndarray
type(sales)
```

```python
#第一种分析方法——基于统计学的建模
```

```python
X_add_const = sm.add_constant(Data.to_numpy())
myModel = sm.OLS(sales, X_add_const)    #采用OLS()方法用最小二乘法来建立myModel模型
results = myModel.fit()    #采用模型的fit()方法返回一个回归结果对象
print(results.summary())
```

```python
#R-squared的值为0.897，接近于1，说明回归效果好。
#F-statistic（F检验）的值越大越能推翻原假设，原假设是“我们的模型不是线性模型”。
#Prob（F-statistic）为1.58e-96,该值非常小，足以证明是线性显著的。
```

```python
#第二种分析方法——基于机器学习
```

```python
#拆分训练集和测试集
X_train, X_test, Y_train, Y_test = train_test_split(Data, sales, random_state = 1, test_size = 0.25)
```

```python
#查看训练集与测试数据的数量
print(X_train.shape)
print(X_test.shape)
```

```python
#调用sklearn.linear_model中默认参数的LinearRegression对训练集进行线性回归
linreg = LinearRegression()
model = linreg.fit(X_train, Y_train)
print(model)
```

```python
model.coef_    #查看多元线性回归模型的回归系数
```

```python
model.intercept_    #查看回归模型的截距
```

```python
model.score(X_test, Y_test)    #准确率
```

```python
#进行预测
Y_pred = linreg.predict(X_test)
Y_pred
```
